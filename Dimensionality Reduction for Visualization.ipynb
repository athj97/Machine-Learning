```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
f = pd.read_csv("lfw20.csv", header = None)
f
```

```python
print(f.to_numpy().shape)
ex = f.to_numpy()[-1, 1:].reshape(87, 65)
ex.shape
```

```python
plt.imshow(ex, cmap = 'gray')
plt.axis('off')
```

```python
data = f.to_numpy()[:, 1:]
index = f.to_numpy()[:, 0]
print(index)
data = data / 255
print(data.shape, data[87, 65])
pd.DataFrame(data).head()
```

```python
zscored_data = (data - np.mean(data, axis = 0))/(np.std(data, axis = 0, ddof = 1))
pd.DataFrame(zscored_data).head()
```

```python
print(type(data), type(zscored_data))
```

```python
w, v = np.linalg.eigh(np.cov(np.transpose(zscored_data)))#'w' represents eigen values and 'v' represents eigen vectors
w, v
```

```python
print(w.shape, v.shape)
#Sorting eigen values in descending order and then using the sorted list for finding 'X'
X = zscored_data @ v[:, w.argsort()[::-1][:2]]
X.shape
```

```python
%matplotlib inline
X = pd.DataFrame(X)
X["index"] = index
for i in X['index'].unique():
    X_temp = X[X['index'] == i]
    plt.scatter(X_temp[0], X_temp[1])
plt.suptitle(t = 'Each color represents a unique set of images', y = 0, ha = 'center')
plt.title('PCA Non-whitened 2D projection')
```

```python
z = v[:, w.argsort()[::-1][:2]]
print(z.shape)
Xw = pd.DataFrame(zscored_data @ (z / np.sqrt(abs(z))))
Xw.head()
```

```python
Xw["index"] = index
for i in Xw['index'].unique():
    X_temp2 = Xw[Xw['index'] == i]
    plt.scatter(X_temp2[0], X_temp2[1])
plt.suptitle(t = 'Each color represents a unique set of images', y = 0, ha = 'center')
plt.title('PCA Whitened 2D projection')
```
